```r
library(tibble)
#library(dplyr)
library(readr)
library(ggplot2)
library(tidyverse)
library(tidyr)
library(MASS)
# Function Estimating Beta Distribution
estBetaParams <- function(mu, var) {
  alpha <- ((1 - mu) / var - 1 / mu) * mu ^ 2
  beta <- alpha * (1 / mu - 1)
  return(params = list(alpha = alpha, beta = beta))
}
# Function to calculate distribution Mode
Modes <- function(x) {
  ux <- unique(x[!is.na(x)])
  tab <- tabulate(match(x, ux))
  ux[tab == max(tab)]}
```
